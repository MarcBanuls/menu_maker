code to make an automated menu
List of things needed:
- List of base ingredients
- List of protein
- List of sauces
- List of exceptions (combinations not possible of base-protein-sauce, eg: curry-quinoa)
- List of dinners
- Rules based on base-protein-sauces between days

Rules:
- No same base two days in a row
- No same protein two days in a row
- No same sauce two days in a row
- No repeat same base-protein-sauce 2 times a week
- No repeat same combination of base-protein-sauce in a week
- possibility of double bas-protein-sauce for curry and hummus
- On a lunch-dinner there should be an equilibrated meat/veggies ratio
- Start on Wednesday
- Make easier to do recipes for Tue Thur
- Prioritize some foods for specific days(pizza-curry on weekends, fajitas on non-telework day)


Creation of lists

```python
#bases proteins sauces
bases = ['Pasta','Quinoa','Rice','Couscous','Potatoes','Tortilla','Lentils','Pizza Crust']
proteins  = ['Chicken','Turkey','Pork','Eggs','Beef','Soy','White fish','Blue fish']
sauces  = ['Tomato','Nata','Japanese Curry','Thai Curry','Veggies','Cheese','Broth']

# Dinners (they go on separate, with full combos)
dinners = ['Traineras','Salad','Hummus','Sandwich','Guacamole','Tortilla','Fish','Pure','Soup','Quiche','Pizzajita','Burguer','Smoothie']
```

```python
#exceptions
#create tuples of exceptions, now only 2 for testing
exceptions = [
    ('Lentils','White fish','Cheese'),
    ('Lentils','Pork','Thai Curry')
]
```

Meal class

```python
class Meal:
    def __init__(self, base=None, protein=None, sauce=None, type="lunch"):
        self.base = base
        self.protein = protein
        self.sauce = sauce
        self.type = type

    def __repr__(self):
        return f"{self.base}-{self.protein}-{self.sauce}"
```

Week Schedule

```python
#Dictionary of days
week_schedule = {
    "Wednesday": {"lunch": None, "dinner": None},
    "Thursday": {"lunch": None, "dinner": None},
    "Friday": {"lunch": None, "dinner": None},
    "Saturday": {"lunch": None, "dinner": None},
    "Sunday": {"lunch": None, "dinner": None},
    "Monday": {"lunch": None, "dinner": None},
    "Tuesday": {"lunch": None, "dinner": None},
}
```

```python
#add valid_lunch function
def valid_lunch(meal, last_meal, week_meals):
    #no repeat same combination
    if last_meal:
        if (meal.base == last_meal.base or meal.protein == last_meal.protein or meal.sauce == last_meal.sauce):
            return False
    # no repeat same combo in a week
    for m in week_meals:
        if (meal.base == m.base and meal.protein == m.protein and meal.sauce == m.sauce):
            return False
        
    # Check if the meal is in exceptions
    if (meal.base, meal.protein, meal.sauce) in exceptions:
        return False
    return True
```

```python
#add valid dinner function #PENDING TO DO
def valid_dinner(dinner,last_dinner, week_dinner):
    # no repeat same dinner in a week
    for m in week_dinner:
        if dinner == m:
            return False
    #no repeat same dinner as last dinner
    if last_dinner:
        if dinner == last_dinner:
            return False
    return True
```

```python
import random

def generate_week_menu():
    last_meal = None
    week_meals = []
    for day, meals in week_schedule.items():
        for meal_type in meals:
            if meal_type == "dinner":
                selected = random.choice(dinners)
                week_schedule[day][meal_type] = selected
                continue
            if meal_type == "lunch":
                valid_options = [
                    Meal(base, protein, sauce)
                    for base in bases
                    for protein in proteins
                    for sauce in sauces
                    if valid_meal(Meal(base, protein, sauce),last_meal,week_meals)
                ]
                
                if valid_options:
                    selected = random.choice(valid_options)
                    week_schedule[day][meal_type] = selected
                    week_meals.append(selected)
                    last_meal = selected
    return week_schedule
```
